# Step 1: Business Understanding

## Dataset
This dataset contains GDP data (current US dollars) for countries from 2000 to 2015. It uses the World Bank's popular GDP indicator (NY.GDP.MKTP.CD).

## Goal
Analyze historical GDP trends and build a model to predict GDP for the year 2016, aiding economic forecasting and decision-making.

## Key Questions
1. Which countries had the most GDP growth from 2000 to 2015?
2. How did GDP behave during the 2008 financial crisis?
3. Can we predict 2016 GDP using past data?
4. What are the regional GDP patterns?
5. Which countries have the fastest long-term growth?


# Step 2: Exploratory Data Analysis (EDA)


```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
data_df = pd.read_csv("Data.csv")

# Step 2: Clean column names for ease of use
data_df.columns = [col.split(' [')[0] if ' [YR' in col else col for col in data_df.columns]

# Step 3: Filter only GDP series
gdp_df = data_df[data_df["Series Code"] == "NY.GDP.MKTP.CD"].copy()

# Step 4: Identify year columns and clean GDP values
year_columns = [col for col in gdp_df.columns if col.isdigit()]
gdp_df[year_columns] = gdp_df[year_columns].replace("..", np.nan).astype(float)

# Step 5: Explore structure and missing values
num_countries = gdp_df.shape[0]
num_years = len(year_columns)

missing_by_year = gdp_df[year_columns].isnull().sum()
min_gdp = gdp_df[year_columns].min().min()
max_gdp = gdp_df[year_columns].max().max()
mean_gdp_by_year = gdp_df[year_columns].mean().sort_values(ascending=False)

# Step 6: Get Top 10 GDP countries in 2015
top_10_2015 = gdp_df.sort_values(by='2015', ascending=False).head(10)[['Country Name', '2015']]

# Display summary
print("Dataset includes GDP data for", num_countries, "countries and", num_years, "years (2000–2015).\n")
print("Missing values in top 5 years with gaps:\n", missing_by_year.sort_values(ascending=False).head(), "\n")
print("GDP ranges from $", round(min_gdp), "to $", round(max_gdp), "USD.\n")
print("Top 5 years by average GDP (most skewed):\n", mean_gdp_by_year.head(), "\n")
print("Top 10 countries by GDP in 2015:\n", top_10_2015.to_string(index=False))
```


## Dataset Structure
- The dataset includes annual GDP (in current US dollars) for 217 countries from 2000 to 2015.
- Each row represents one country, and each column represents the GDP in a specific year.
- The source of the data is the World Bank's "NY.GDP.MKTP.CD" indicator.

## Missing Data
- Most countries have data for all 16 years.
- Missing entries are minimal (e.g., 13 missing in 2000, only 6 by 2015).

## GDP Value Distribution (2000–2015)
- There's a wide gap between the richest and poorest countries.
- GDP values range from as low as 15 milion to over 14 trillion USD$.
- The data is right-skewed due to economic giants like the US and China.

## Top 10 Countries by GDP in 2015


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Plot top 10 GDPs in 2015
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_2015, x='2015', y='Country Name', palette='viridis')
plt.title('Top 10 Countries by GDP in 2015')
plt.xlabel('GDP in Current US Dollars')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# Visualize GDP trend for a few major economies over time
major_countries = ['United States', 'China', 'Japan', 'Germany', 'India']
plt.figure(figsize=(14, 7))
for country in major_countries:
    row = gdp_df[gdp_df['Country Name'] == country]
    plt.plot(year_columns, row[year_columns].values.flatten(), label=country)

plt.title('GDP Trend (2000–2015)')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## GDP Trend for Major Economies

The plot above shows GDP growth over time for five major economies:
- **United States** maintains the highest GDP across all years.
- **China** shows the fastest growth rate, nearly doubling every few years.
- **India** , **Japan** and **Germany** also show steady growth.
- This analysis gives us confidence that GDP follows strong temporal trends, which is suitable for prediction modeling.
